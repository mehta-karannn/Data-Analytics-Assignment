```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
```

```python
##Create a Line Plot
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic line plot')
plt.show()
```

```python
##  Customize line plot

plt.plot(x,y,color="red",linestyle='--',linewidth=6)
plt.title("Customize Line plot ")
plt.show()
```

```python
##Marker and marker size and grid

plt.plot(x,y,marker="o",markersize=9)
plt.grid(True)
```

```python
##Multiple plots
x=[1,2,3,4,5]
y=[1,4,9,16,25]
z=[1,8,27,64,125]

plt.figure(figsize=(9,5))
plt.subplot(3,2,1)
plt.plot(x,y,color="Red")
plt.title("Hi")
plt.subplot(3,2,2)
plt.plot(x,z,color="Blue")
plt.title("Plot 2")
plt.subplot(3,2,3)
plt.plot(y,z,color="green")
plt.title("plot 3")
plt.subplot(3,2,4)
plt.plot(y,z,color="yellow")
plt.title("karan")
```

```python
##bar Plot
Names=['Karan','Tanvi','Kajol','Husein','Faizan','Reva','Tanvi R','Hamza']
Marks=[20,20,18,14,19,10,14,20]

plt.bar(Names,Marks,color='Red')
plt.xlabel("Names")
plt.ylabel('Marks')
plt.title('Marks Of EVM subject')
plt.show()
```
