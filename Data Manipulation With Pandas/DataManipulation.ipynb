```python
## FillNA function!!
## if we are doing fillna with 0 then always make a copy data of it never do it in orignal data
data_filled=data.fillna(0)
data_filled
```

```python
## we usually take mean of the data 
data['Sale_Manu']=data['Sales'].fillna(data['Sales'].mean())
data
```

```python
##to rename the column name
data=data.rename(columns={"Date":"Sales Date"})
data.head(5)
```

```python
data["Value_new"]=data["Value"].fillna(data["Value"].mean()).astype(int)
data
```

```python
def square(x):
    return x*3

data["New Value"]=data["Value"].apply(square)
data
```

```python
###Data Aggregating and grouping
grouped_mean=data.groupby('Product')["Value"].mean()
grouped_mean
```

```python
aggreated_summary=data.groupby(['Region','Product','Sales'])["Value"].agg(["mean","sum","count"])
aggreated_summary
```

```python
##Merging and Joining
df1=pd.DataFrame({"Name":["Karan","Tanvi","Husien","Kajol"],"Roll_No":[35,10,41,4]})
df2=pd.DataFrame({"Roll_No":[35,10,41,4,32],"Marks":[19,19,20,20,10]})

print(df1)
print(df2)
```

```python
pd.merge(df1,df2,on="Roll_No",how="inner")
```

```python
pd.merge(df1,df2,on="Roll_No",how="outer")
```

```python
pd.merge(df1,df2,on="Roll_No",how="left")
```

```python
pd.merge(df1,df2,on="Roll_No",how="right")
```
